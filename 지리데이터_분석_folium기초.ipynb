# 3. 마커와 팝업 사용하기
## 3.1. 단일 마커 추가하기
folium을 사용하여 지도에 마커를 추가하는 것은 간단합니다. Marker 클래스를 사용하여 특정 위도와 경도에 마커를 추가할 수 있습니다.

```python
import folium

# 서울시청을 중심으로 하는 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 마커 추가
folium.Marker([37.5665, 126.9780], tooltip='서울시청').add_to(m)

m
```

## 3.3. 여러 마커를 그룹으로 묶어서 표시하기 (MarkerCluster)

여러 위치에 마커를 추가할 때, 지도가 확대/축소될 때마다 적절한 수준의 마커를 그룹으로 묶어 표시하려면 MarkerCluster를 사용할 수 있습니다. 이 기능을 사용하기 위해선 folium.plugins에서 MarkerCluster를 임포트해야 합니다.

```python
from folium.plugins import MarkerCluster

# 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 마커 클러스터 생성
marker_cluster = MarkerCluster().add_to(m)

# 여러 위치에 마커 추가
locations = [
    [37.5665, 126.9780],
    [37.5776, 126.9768],
    [37.5592, 126.982],
    [37.5653, 126.9896]
]

for location in locations:
    folium.Marker(location).add_to(marker_cluster)

m
```

이 예제는 서울시청 주변의 여러 위치에 마커를 추가하고, 이 마커들을 클러스터로 그룹화합니다. 지도의 확대/축소 수준에 따라 마커가 자동으로 그룹화되거나 개별 표시됩니다.

이러한 방법들을 통해 folium을 사용하여 지도에 다양한 마커와 정보를 표시할 수 있습니다.

# 4. 지도에 다양한 형태 추가하기
## 4.1. 원(Circle)과 원 마커(CircleMarker) 추가하기

Circle과 CircleMarker는 지도에 원 형태로 정보를 표시하는 데 사용됩니다. 두 형태의 주요 차이는 확대/축소 시 크기 변화 방식입니다. Circle은 지구의 실제 면적에 따라 크기가 고정되어 있습니다. 반면, CircleMarker는 화면에서의 크기가 고정되어 있습니다.

```python
import folium

# 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 원 추가
folium.Circle(
    location=[37.5665, 126.9780],
    radius=2000,  # 미터 단위
    color='blue',
    fill=True,
    fill_color='blue'
).add_to(m)

# 원 마커 추가
folium.CircleMarker(
    location=[37.5765, 126.9880],
    radius=30,  # 픽셀 단위
    color='red',
    fill=True,
    fill_color='red'
).add_to(m)

m
```

## 4.2. 폴리라인(PolyLine) 및 다각형(Polygon) 그리기
여러 위치를 연결하여 선이나 다각형을 그릴 수 있습니다.

```python
# 폴리라인 추가
coords = [[37.5665, 126.9780], [37.5765, 126.9880], [37.5665, 126.9980]]
folium.PolyLine(coords, color='green').add_to(m)

# 다각형 추가
coords_polygon = [[37.5565, 126.9780], [37.5465, 126.9880], [37.5565, 126.9980]]
folium.Polygon(coords_polygon, color='yellow', fill=True, fill_color='yellow').add_to(m)

m
```

## 4.3. 직사각형(Rectangle) 및 원형(Circle) 영역 표시하기
지도에 특정 영역을 표시하기 위해 직사각형이나 원을 사용할 수 있습니다.

```python
# 직사각형 추가
bounds = [[37.5665, 126.9680], [37.5765, 126.9880]]
folium.Rectangle(bounds, color='purple', fill=True, fill_color='purple').add_to(m)

# 원형 영역 표시는 위에서 'Circle'을 사용하여 이미 설명했습니다.

m
```

이러한 예제를 통해 folium을 사용하여 지도에 다양한 형태와 정보를 시각적으로 표시하는 방법을 알게 되었습니다.

# 5. 지도에 GeoJSON과 TopoJSON 레이어 추가하기
## 5.1. GeoJSON 및 TopoJSON 데이터 이해하기
GeoJSON과 TopoJSON은 지리적 객체를 나타내기 위한 JSON 포맷입니다.

[GeoJSON]

웹에서 지리적 데이터를 교환하는 데 널리 사용되는 포맷입니다.
주로 지리적 특성의 컬렉션으로, 각 특성은 기하학적 객체와 그와 연결된 속성을 가집니다.
기하학적 객체 유형에는 "Point", "LineString", "Polygon" 등이 있습니다.

[TopoJSON]

GeoJSON의 확장된 형태입니다.
연결된 지리적 특성 간의 중복을 제거하여 파일 크기를 줄입니다.
주로 대규모 지리 데이터를 웹에서 효율적으로 전송하고 사용할 때 적합합니다.

## 5.2. folium을 사용하여 GeoJSON 및 TopoJSON 레이어 추가하기
folium을 사용하면 GeoJSON 및 TopoJSON 데이터를 쉽게 지도에 추가할 수 있습니다.

```python
geo_data = {
    "type": "Feature",
    "geometry": {
        "type": "Point",
        "coordinates": [126.9780, 37.5665]
    },
    "properties": {
        "name": "서울시청"
    }
}
```

이 GeoJSON 데이터는 "서울시청"의 지리적 위치를 나타냅니다.

GeoJSON 레이어를 folium 지도에 추가하는 코드는 다음과 같습니다.

```python
import folium

# 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# GeoJSON 레이어 추가
folium.GeoJson(geo_data, name="서울시청").add_to(m)

m
```

TopoJSON 데이터도 유사한 방법으로 추가할 수 있지만, folium은 직접적으로 TopoJSON 형식을 지원하지 않습니다. TopoJSON을 사용하려면 추가 처리가 필요하며, 외부 라이브러리를 활용하거나 데이터를 GeoJSON 형식으로 변환한 후 사용하는 것이 일반적입니다.

# 6. Choropleth 지도 만들기
## 6.1. Choropleth 지도의 개념 및 사용 사례

[Choropleth 지도의 개념]

Choropleth 지도는 지리적 영역(예: 국가, 주, 도시 등)의 경계 내에서 어떤 수량적 변수의 값을 색상 또는 패턴의 차이로 나타내는 지도입니다. 이러한 지도는 주어진 지역의 특정 값 또는 속성의 분포와 패턴을 시각적으로 표현하는 데 유용합니다.

* 인구 밀도: 국가별, 주별 또는 도시별 인구 밀도를 나타내기 위해 사용됩니다.
* 선거 결과: 각 지역에서 어떤 후보나 정당이 얼마나 많은 표를 얻었는지 보여줍니다.
* 소득 수준: 지역별 평균 소득이나 경제 지표를 나타냅니다.
* 환경 데이터: 지역별 오염 농도나 기후 지표를 표시합니다.

```python
import folium
import numpy as np

# 기준 위치 설정 (서울시 중심)
base_location = [37.5665, 126.9780]
```

```python
num_shops = 100

# 무작위로 상점 위치와 매출 생성
np.random.seed(42)  # 일관된 결과를 위해 seed 설정
lats = base_location[0] + np.random.rand(num_shops) * 0.05
lngs = base_location[1] + np.random.rand(num_shops) * 0.05
sales = np.random.randint(100000, 100000000, num_shops)

shops_data = list(zip(lats, lngs, sales))
shops_data[:5]  # 처음 5개 상점의 데이터를 출력합니다.
```

```python
# 기본 지도 생성
m = folium.Map(location=base_location, zoom_start=13)

# 매출에 따른 마커 색상 설정 함수
def get_color(sale):
    if sale > 80000000:
        return 'red'
    elif sale > 50000000:
        return 'orange'
    else:
        return 'blue'

# 상점 데이터를 지도에 추가
for lat, lng, sale in shops_data:
    folium.Marker(
        location=[lat, lng],
        icon=folium.Icon(color=get_color(sale)),
        popup=f"매출: {sale:,}원"
    ).add_to(m)

m
```

```python
# 격자 크기 설정
grid_size = 0.01  # 0.01도

# 지역의 경계 설정
lat_min, lat_max = base_location[0] - 0.05, base_location[0] + 0.05
lng_min, lng_max = base_location[1] - 0.05, base_location[1] + 0.05
```

```python
# 격자 크기에 따른 격자 수 계산 (1을 추가하여 경계를 포함)
num_lat_grids = int((lat_max - lat_min) // grid_size) + 1
num_lng_grids = int((lng_max - lng_min) // grid_size) + 1

# 격자별 상점 수와 총 매출 초기화
shop_counts = np.zeros((num_lat_grids, num_lng_grids))
total_sales = np.zeros_like(shop_counts)

# 각 상점 데이터를 격자에 할당
for lat, lng, sale in shops_data:
    i = int((lat - lat_min) // grid_size)
    j = int((lng - lng_min) // grid_size)
    shop_counts[i, j] += 1
    total_sales[i, j] += sale

# 격자별 평균 매출 계산 (0으로 나누는 것을 방지하기 위해 np.where 사용)
avg_sales = np.where(shop_counts != 0, total_sales / shop_counts, 0)

shop_counts, avg_sales
```

```python
# 밀집도에 따른 지도 생성
density_map = folium.Map(location=base_location, zoom_start=13)

# 격자의 색상 결정 함수
def get_density_color(count):
    if count > 5:
        return 'red'
    elif count > 2:
        return 'orange'
    elif count > 0:
        return 'yellow'
    else:
        return 'transparent'

for i in range(num_lat_grids):
    for j in range(num_lng_grids):
        # 격자의 중심점 계산
        center_lat = lat_min + i * grid_size + grid_size / 2
        center_lng = lng_min + j * grid_size + grid_size / 2

        # 격자 표시
        folium.Rectangle(
            bounds=[[lat_min + i * grid_size, lng_min + j * grid_size],
                    [lat_min + (i + 1) * grid_size, lng_min + (j + 1) * grid_size]],
            color=get_density_color(shop_counts[i, j]),
            fill=get_density_color(shop_counts[i, j]),
            fill_opacity=0.4,
            popup=f"상점 수: {int(shop_counts[i, j])}"
        ).add_to(density_map)

density_map
```

```python
# 평균 매출에 따른 지도 생성
sales_map = folium.Map(location=base_location, zoom_start=13)

# 평균 매출에 따른 격자의 색상 결정 함수
def get_sales_color(avg_sale):
    if avg_sale > 60000000:
        return 'red'
    elif avg_sale > 40000000:
        return 'orange'
    elif avg_sale > 0:
        return 'yellow'
    else:
        return 'transparent'

for i in range(num_lat_grids):
    for j in range(num_lng_grids):
        # 격자의 중심점 계산
        center_lat = lat_min + i * grid_size + grid_size / 2
        center_lng = lng_min + j * grid_size + grid_size / 2

        # 격자 표시
        folium.Rectangle(
            bounds=[[lat_min + i * grid_size, lng_min + j * grid_size],
                    [lat_min + (i + 1) * grid_size, lng_min + (j + 1) * grid_size]],
            color=get_sales_color(avg_sales[i, j]),
            fill=get_sales_color(avg_sales[i, j]),
            fill_opacity=0.4,
            popup=f"평균 매출: {int(avg_sales[i, j]):,}원"
        ).add_to(sales_map)

sales_map
```
