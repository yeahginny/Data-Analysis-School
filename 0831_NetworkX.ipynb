```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node("A")
G.add_node("B")
```

```python
# 그래프 시각화
nx.draw(G, with_labels=True)
plt.show()
```

```python
G.add_edge("A", "B")
```

```python
# 그래프 시각화
nx.draw(G, with_labels=True)
plt.show()
```

```python
#nx.Graph 기본 사용법

G = nx.Graph()

G.add_node(1) # 노드 추가 (1)
G.add_nodes_from([2,3]) # 여러개의 노드 추가 [2,3]
```

```python
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2, 3])
G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])
```

```python
# 그래프 시각화
nx.draw(G, with_labels=True)
plt.show()
```

```python
node_colors = [G.nodes[node].get("color", "blue") for node in G.nodes()]
```

```python
node_colors
```

```python
# 그래프 시각화
plt.figure(figsize=(7, 7))
nx.draw(G, with_labels=True, node_color=node_colors, node_size=1000, edge_color='gray', font_size=15)
plt.title("Graph with Node Attributes")
plt.show()
```

```python
H = nx.path_graph(20)
```

```python
# 그래프 시각화
nx.draw(H, with_labels=True)
plt.show()
```

```python
G.add_nodes_from(H)
```

```python
# 노드 속성에 따른 색상 지정 (기존 코드 사용)
node_colors = [G.nodes[node].get("color", "blue") for node in G.nodes()]
```

```python
# 그래프 시각화
plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_color=node_colors, node_size=1000, edge_color='gray', font_size=15)
plt.title("Extended Graph with Path Graph Nodes")
plt.show()
```

```python
e = (1,3)
G.add_edge(*e)

# 노드 속성에 따른 색상 지정 (이전 코드에서 사용된 것을 재사용)
node_colors = [G.nodes[node].get("color", "blue") if not isinstance(node, nx.Graph) else "yellow" for node in G.nodes()]

# 그래프 시각화
plt.figure(figsize=(10, 10))
nx.draw(G, with_labels=True, node_color=node_colors, node_size=1000, edge_color='gray', font_size=15)
plt.title("Graph with Additional Edges")
plt.show()
```

```python
G.clear()
```

```python
# 그래프 시각화
nx.draw(G, with_labels=True)
plt.show()
```

```python
G = nx.Graph()
```

```python
G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')
```

```python
# 그래프 시각화
nx.draw(G, with_labels=True)
plt.show()
```
